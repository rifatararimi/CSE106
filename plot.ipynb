```python
# importing the required module
import matplotlib.pyplot as plt
```

```python
# x axis values
ct = [0.002,0.004,0.004,0.005,0.015]
# corresponding y axis values
d = [10,20,40,80,160]

# plotting the points
plt.plot(d,ct)

# naming the x axis
plt.xlabel('Distance')
# naming the y axis
plt.ylabel('Computional Time (s)')

# giving a title to my graph
plt.title('Computional Time vs Distance')

# function to show the plot
plt.show()
```
